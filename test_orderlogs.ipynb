```python
import pandas as pd
from pprint import pprint
```

```python
filename = "log_files/order7_13.log"
```

```python
orders = []
with open(filename,"r") as f:
    for line in f:
        if ("response" not in line):
            print(line)
            orders.append(line)
```

```python
orderdict_list = []
for line in orders:
    print(line)
    split_line = line.split(";")
    print(split_line)
    order_dict = {}
    order_dict["timestamp"] = line[11:34]
    order_dict["eid"] = int(str(split_line[1])[4:])
    order_dict["type"] = str(split_line[2])[5:]
    order_dict["order"] = str(split_line[3])[6:]
    order_dict["asset"] = str(split_line[4])[6:]
    order_dict["price"] = int(str(split_line[5])[6:])
    qty = str(split_line[6])[4:]
    if qty == "null":
        order_dict["qty"] = qty
    else:
        order_dict["qty"] = int(qty)
    order_dict["message"] = str(line.split("message:")[1])[:-1]
    print(order_dict)
    orderdict_list.append(order_dict)
```

```python
orderdict_list = []
for line in orders:
    order_dict = {}
    if "Buying" in line:
        print(line)
        order_dict["timestamp"] = line[11:34]
        split_line = line[34:].split(" ")
        print(split_line)
        order_dict["eid"] = int(str(split_line[0])[1:-1])
        order_dict["side"] = split_line[2]
        order_dict["asset"] = split_line[4]
        order_dict["price"] = int(str(split_line[5])[3:])
        order_dict["qty"] = int(str(split_line[6])[3:])
        order_dict["message"] = str(line.split("message:")[1])[:-1]
        orderdict_list.append(order_dict)
        # pprint(order_dict)
    elif "Selling" in line:
        print(line)
        order_dict["timestamp"] = line[11:34]
        split_line = line[34:].split(" ")
        print(split_line)
        order_dict["eid"] = int(str(split_line[0])[1:-1])
        order_dict["side"] = split_line[2]
        order_dict["asset"] = split_line[3]
        order_dict["price"] = int(str(split_line[4])[3:])
        print(str(split_line[5]))
        print(str(split_line[5])[4:])
        order_dict["qty"] = int(str(split_line[5])[3:])
        order_dict["message"] = str(line.split("message:")[1])[:-1]
        orderdict_list.append(order_dict)
        # pprint(order_dict)
    elif "cancel" in line:
        print(line)
        order_dict["timestamp"] = line[11:34]
        split_line = line[34:].split(" ")
        order_dict["eid"] = int(str(split_line[0])[1:-1])
        order_dict["side"] = split_line[1] + split_line[2]
        order_dict["asset"] = split_line[6]
        order_dict["price"] = split_line[7]
        order_dict["qty"] = "X"
        order_dict["message"] = str(line.split("message:")[1])[:-1]
        orderdict_list.append(order_dict)
    else:
        print(line)
```

```python
orders_df = pd.DataFrame(orderdict_list)
```

```python
filename = "log_files/order7_13.log"
```

```python
orders = []
with open(filename,"r") as f:
    for line in f:
        if ("response" not in line):
            print(line)
            orders.append(line)
```

```python
orderdict_list = []
for line in orders:
    order_dict = {}
    if 1==1:
        print("line: ",line[:-1])
        split_line = line.split(";")
        print("split: ",split_line)
        order_dict["timestamp"] = split_line[0][-23:]
        order_dict["eid"] = int(split_line[1][4:])
        order_dict["type"] = split_line[2][5:]
        order_dict["side"] = split_line[3][6:]
        order_dict["asset"] = split_line[4][6:]
        order_dict["price"] = int(split_line[5][6:])
        qty_value = split_line[6][4:]
        if qty_value == "null":
            order_dict["qty"] = qty_value
        else:
            order_dict["qty"] = int(qty_value)
        order_dict["message"] = split_line[7][8:][:-1]
        # order_dict["orderid"] = response["orderid"]
        orderdict_list.append(order_dict)
        # pprint(order_dict)
        # print("dict: ",order_dict)
    else:
        print(line)
```

```python
orders_df = pd.DataFrame(orderdict_list)
orders_df[["m1","m2","m3","m4"]] = orders_df["message"].str.split(',',expand=True)
df1 = orders_df[orders_df["eid"]==12400]
```
