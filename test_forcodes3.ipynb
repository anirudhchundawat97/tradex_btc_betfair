```python
from event_data_static import EventDataStatic
from pprint import pprint
import pandas as pd
from tqdm.notebook import tqdm
```

```python
eds = EventDataStatic("p",603727)
```

```python
im_list = eds.get_live_im_event_ids()
```

```python
event_det_list = []
for i in tqdm(im_list):
    all_data = eds.get_event_det_from_id(i)
    keys = ['auto_match', 'category','start_date', 'endsat', 'max_allowed_position', "nCP", "yCP", "subcategory", "title"]
    relv_data = {x:all_data[x] for x in keys}
    relv_data["eid"] = i
    # pprint(all_data)
    event_det_list.append(relv_data)
```

```python
event_det_df = pd.DataFrame(event_det_list)
```

```python
event_det_df["price_diff"] = abs(event_det_df["yCP"] - event_det_df["nCP"])
```

```python
event_det_df.sort_values(by="price_diff")
```

```python
event_det_df["category"].unique()
```

```python
event_det_df[event_det_df["category"]=="Finance"].sort_values(by="price_diff")
```
