```python
eid_list = range(23034, 23130)
total = 0

print("Event", "PnLFinal", "PnlWorst", "Settled", "        Qty")
print("-"*45)


for i in eid_list:
    txn = Transactions(i, "p", 603727)
    txn.update()
    
    if (txn.event_txn_df is None) or (txn.event_txn_df.empty):
        pnl = np.nan
        qty = np.nan
        pnl_worst = np.nan
    else:
        pnl = txn.event_txn_df["amount"].iloc[:].sum()
        qty = txn.event_txn_df["qty"].iloc[:].sum()
        try:
            db_df = sr.get_data_df_from_psql_live1(i)
            pnl_worst = db_df["pnl_worst"].min()
        except:
            pnl_worst = None
        
    if "Settled" in txn.event_txn_df["status"].values:
        settle = "settled"
    elif (np.isnan(pnl)) and (np.isnan(qty)) and (np.isnan(pnl_worst)):
        settle = np.nan
    else:
        settle = "not settled"
    print(i, f"{pnl:>8}", f"{pnl_worst:>8}", f"{settle:<11}", f"{qty:>6}")
    if not np.isnan(pnl):
        total = total + pnl
print("-"*45)
print("total final pnl: ", total)
```
